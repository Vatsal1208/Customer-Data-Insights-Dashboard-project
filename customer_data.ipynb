```python
orders_csv = "datasets/List of Orders.csv"
order_details_csv ="datasets/Order Details.csv"
sales_target_csv = "datasets/Sales target.csv"


print("Files will be loaded from:")
print(orders_csv)
print(order_details_csv)
print(sales_target_csv)
```

```python
orders_df = pd.read_csv(orders_csv)
order_details_df = pd.read_csv(order_details_csv)
sales_targets_df = pd.read_csv(sales_target_csv)

print("Orders Data:", orders_df.head())

print("Order Details Data:", order_details_df.head())

print("Sales Targets Data:", sales_targets_df.head())
```

```python
orders_df.info()
```

```python
orders_df.drop_duplicates(inplace=True)
order_details_df.drop_duplicates(inplace=True)
sales_targets_df.drop_duplicates(inplace=True)

orders_df["Order Date"] = pd.to_datetime(orders_df["Order Date"], dayfirst=True)
```

```python
# orders_df.isnull().sum()
# order_details_df.isnull().sum()
```

```python
orders_df.dropna()
```

```python
join_df = pd.merge(orders_df, order_details_df, on="Order ID", how="inner")
print(join_df)
```

```python
print("Missing values in Orders:")
print(orders_df.isnull().sum())
print(orders_df)

print("\nMissing values in Order Details:")
print(order_details_df.isnull().sum())
```

Combined CSV File after cleaning data with multiple orders also 


```python
combined_csv = os.path.join("customer_data_insights", "combined.csv")
os.makedirs("customer_data_insights", exist_ok=True)

if not os.path.exists(combined_csv):
    join_df.to_csv(combined_csv, index=False)
    print(f"File created and saved to: {combined_csv}")
else:
    print(f"File already exists, loading: {combined_csv}")

combined_df = pd.read_csv(combined_csv)
```
